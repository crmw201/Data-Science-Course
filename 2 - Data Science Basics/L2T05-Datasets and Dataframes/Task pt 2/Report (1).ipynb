```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

```python
# Compare average income based on ethnicity 
#print("The average income of Caucasian is", df[df.Ethnicity == "Caucasian"].loc[:,"Income"])
#print("The average income of Asian is", df[df.Ethnicity == "Asian"].loc[:,"Income"])
#print("The average income of African American is", df[df.Ethnicity == "African American"].loc[:,"Income"])

print("Average income by ethnicity:")
print(df.groupby("Ethnicity")["Income"].mean())
```

```python
# On average, do married or single people have a higher balance?

print("Average balance by relatiionship status:")
print(df.groupby("Married")["Balance"].mean())
```

```python
# What is the highest income in our dataset?
print("The highest income in our dataset is", df.loc[:, "Income" ].max())
```

```python
# What is the lowest income in our dataset?
print("The highest income in our dataset is", df.loc[:, "Income" ].min())
```

```python
# How many cards do we have recorded in our dataset?
print("The total number of cards in our dataset is", df["Cards"].sum())
```
