```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load data
df = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
```

### 1. Boxplot

```python
# Boxplot for revs per mile for Audi, Hyundai, Suzuki and Toyota
selected_manufacturers = ["Audi", "Hyundai", "Suzuki", "Toyota"]
filtered_df = df[df['Manufacturer'].isin(selected_manufacturers)]

plt.figure(figsize=(10,6))
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=filtered_df)
plt.title('Boxplot of Revs per mile for Selected Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Rev.per.mile')
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**


Suzuki

### 2. Histogram

```python
# Histogram of MPG
plt.figure()
plt.hist(df['MPG.city'], color='blue', label='City')
plt.hist(df['MPG.highway'], color='orange', label='Highway')
plt.title('Comparison of MPG in City vs Highway')
plt.xlabel('Miles per Gallon')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**

Generally more fuel efficient to drive on the highway.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure()
sns.lineplot(x='Wheelbase', y='Turn.circle', data=df)
plt.show()
plt.savefig('sns_lineplot.png')
plt.close()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  


Positive relationship, with positive coefficient. Turncircle gets larger as wheelbase gets larger.

### 4. Barplot

```python
# Barplot for mean horsepower

# Calculate mean horsepower by type
mean_horsepower = df.groupby('Type')['Horsepower'].mean()

plt.figure()
mean_horsepower.plot(kind='bar', color='blue')
plt.title('Barplot for Mean Horsepower by Car Type')
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**

Looking at Small and Compact as our smaller vehicles, we can say that Large,
Midsize and Van all exhibit the case that a larger car generally means more 
horsepower.
