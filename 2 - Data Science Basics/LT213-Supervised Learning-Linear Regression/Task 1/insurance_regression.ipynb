```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

Task: Use the data in the relevant columns from the insurance data
to determine how age affects insurance costs.

1. Plot a scatter plot with age on the x-axis and charges on the y-axis

```python
# Import the data
insurance_df = pd.read_csv("insurance.csv")
insurance_df.head()
```

```python
x = insurance_df.iloc[:,:1].values
y = insurance_df.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

2. Using linear_model.LinearRegression() from sklearn, fit a model to your data,
and make predictions on the data.

```python
# Reshape to 2D array, needed by sklearn

X = insurance_df['age'].values.reshape(-1, 1)
y = insurance_df['charges'].values
```

```python
# Fit the model and predict

insurance_model = LinearRegression()
insurance_model.fit(X,y)
y_pred = insurance_model.predict(X)
```

```python
# Plot model and data

plt.figure()
sns.scatterplot(data=insurance_df, x='age', y='charges', color='red')
plt.title("Simple Linear Regression: Charges vs Age")
plt.plot(X, y_pred)
plt.show()
plt.close()
```

```python
# Model Coefficients and Intercept

print("Model Coefficient is: ", insurance_model.coef_)
print("Model intercept is: ", insurance_model.intercept_)
```
